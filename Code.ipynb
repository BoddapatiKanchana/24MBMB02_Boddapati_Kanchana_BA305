```python

#loading the dataset
spotify_df = spark.table("workspace.default.spotify_songs")
lyrics_df  = spark.table("workspace.default.lyrics_10_k")

print("Spotify rows:", spotify_df.count())
print("Lyrics rows:", lyrics_df.count())

display(spotify_df.limit(5))
display(lyrics_df.limit(5))
```

```python
#Data cleaning
# Rename columns to prevent duplication
spotify_df = (spotify_df
              .withColumnRenamed("track_name", "spotify_song")
              .withColumnRenamed("track_artist", "spotify_artist")
              .withColumnRenamed("lyrics", "spotify_lyrics"))

lyrics_df = (lyrics_df
             .withColumnRenamed("song", "lyrics_song")
             .withColumnRenamed("artists", "lyrics_artist")
             .withColumnRenamed("lyrics", "lyrics_text"))

# Join safely on song and artist
combined_df = spotify_df.join(
    lyrics_df,
    (lower(spotify_df.spotify_song) == lower(lyrics_df.lyrics_song)) |
    (lower(spotify_df.spotify_artist) == lower(lyrics_df.lyrics_artist)),
    "left"
).dropDuplicates(["spotify_song", "spotify_artist"])

# Merge text columns safely
combined_df = (combined_df
               .withColumn("lyrics", coalesce(col("lyrics_text"), col("spotify_lyrics")))
               .withColumn("song", coalesce(col("lyrics_song"), col("spotify_song")))
               .withColumn("artist", coalesce(col("lyrics_artist"), col("spotify_artist"))))

# Drop extra columns to remove ambiguity
cols_to_drop = ["spotify_lyrics", "lyrics_text", "lyrics_song", "spotify_song", "lyrics_artist", "spotify_artist"]
for c in cols_to_drop:
    if c in combined_df.columns:
        combined_df = combined_df.drop(c)

print("✅ Combined dataset ready. Columns:", combined_df.columns)

display(combined_df.limit(5))
```

```python
#USE CASE 1 SONG POPULARITY PREDICTION
from pyspark.ml.feature import VectorAssembler
from pyspark.ml.regression import LinearRegression
from pyspark.ml.evaluation import RegressionEvaluator

# Select features
feature_cols = ['danceability', 'energy', 'valence', 'acousticness',
                'instrumentalness', 'liveness', 'tempo']

popularity_df = combined_df.select(*feature_cols, "track_popularity").na.drop()
assembler = VectorAssembler(inputCols=feature_cols, outputCol="features")
data = assembler.transform(popularity_df).select("features", col("track_popularity").alias("label"))

# Split
train, test = data.randomSplit([0.8, 0.2], seed=42)

# Model
lr = LinearRegression(featuresCol="features", labelCol="label")
lr_model = lr.fit(train)
predictions = lr_model.transform(test)

# Evaluate
r2 = RegressionEvaluator(labelCol="label", predictionCol="prediction", metricName="r2").evaluate(predictions)
print(f"🎯 Model R² Score: {r2:.3f}")

# Visualization
coeffs = lr_model.coefficients.toArray()
plt.figure(figsize=(8,4))
plt.barh(feature_cols, coeffs, color='skyblue')
plt.title("Feature Importance for Song Popularity")
plt.xlabel("Coefficient Value")
plt.grid(True)

# Convert to Pandas for visualization
coeff_df = pd.DataFrame({"Feature": feature_cols, "Coefficient": coeffs})

# Use Databricks built-in chart
display(coeff_df)

```

```python
#USE CASE 2 SENTIMENT ANALYSIS OF LYRICS 
# Define UDF for sentiment polarity
def get_sentiment(text):
    try:
        return float(TextBlob(text).sentiment.polarity)
    except:
        return 0.0

sent_udf = udf(get_sentiment, DoubleType())

# Apply on unified lyrics column
sentiment_df = combined_df.withColumn("sentiment_score", sent_udf(col("lyrics")))

# Classify sentiment
sentiment_df = sentiment_df.withColumn(
    "sentiment_label",
    when(col("sentiment_score") > 0.1, "Positive")
    .when(col("sentiment_score") < -0.1, "Negative")
    .otherwise("Neutral")
)

# Sentiment distribution
sent_summary = sentiment_df.groupBy("sentiment_label").count().toPandas()

# Visualization
plt.figure(figsize=(6,6))
plt.pie(sent_summary['count'], labels=sent_summary['sentiment_label'],
        autopct='%1.1f%%', colors=['#90EE90', '#FFD700', '#FF6347'])
plt.title("Lyrics Sentiment Distribution")

sent_summary = sentiment_df.groupBy("sentiment_label").count()
display(sent_summary)

```

```python
#USE CASE 3 GENRE BASED SONG PROFILING
genre_df = (sentiment_df.groupBy("playlist_genre")
    .agg(
        F.avg("danceability").alias("avg_danceability"),
        F.avg("energy").alias("avg_energy"),
        F.avg("valence").alias("avg_valence"),
        F.avg("tempo").alias("avg_tempo")
    )
    .orderBy("avg_energy", ascending=False)
    .na.drop())

genre_pd = genre_df.toPandas()

# Visualization
plt.figure(figsize=(10,5))
plt.bar(genre_pd['playlist_genre'], genre_pd['avg_energy'], color='cornflowerblue')
plt.xticks(rotation=45, ha='right')
plt.title("Average Energy by Genre")
plt.xlabel("Genre")
plt.ylabel("Average Energy")
plt.tight_layout()

genre_df = (sentiment_df.groupBy("playlist_genre")
            .agg(F.avg("energy").alias("avg_energy"))
            .orderBy("avg_energy", ascending=False))
display(genre_df)

```

```python
#USE CASE 4 ARTIST PERFORMANCE & DIVERSTIY ANALYSIS
artist_df = (sentiment_df.groupBy("artist")
    .agg(
        F.avg("track_popularity").alias("avg_popularity"),
        F.avg("sentiment_score").alias("avg_sentiment"),
        F.stddev("tempo").alias("tempo_diversity"),
        F.countDistinct("song").alias("song_count")
    )
    .orderBy(F.col("avg_popularity").desc())
    .na.drop())

top_artists_pd = artist_df.limit(10).toPandas()

# Visualization
plt.figure(figsize=(10,5))
plt.barh(top_artists_pd['artist'], top_artists_pd['avg_popularity'], color='teal')
plt.title("Top 10 Artists by Average Popularity")
plt.xlabel("Average Popularity")
plt.ylabel("Artist")
plt.gca().invert_yaxis()

artist_df = (sentiment_df.groupBy("artist")
             .agg(F.avg("track_popularity").alias("avg_popularity"))
             .orderBy(F.col("avg_popularity").desc())
             .limit(10))
display(artist_df)

```

```python
# ---------------------------
# Use Case 5 — Serverless-safe clustering using scikit-learn on a sampled Pandas DF
# ---------------------------

# 1) Install scikit-learn (run once in a notebook cell; restart kernel if required)
# %pip install scikit-learn

# 2) Imports
from pyspark.sql import functions as F
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.cluster import MiniBatchKMeans

# 3) Feature list (must exist in sentiment_df)
cluster_features = ['danceability','energy','valence','acousticness','tempo','sentiment_score']

# 4) Prepare a cleaned Spark DF for sampling (drop rows with nulls in these features)
spark_cluster_df = sentiment_df.select(*cluster_features).na.drop()

# 5) Decide sample size (safe small sample for driver)
MAX_ROWS = 3000   # maximum rows to move to driver (adjust down if you still hit memory issues)
TOTAL = spark_cluster_df.count()
fraction = min(1.0, float(MAX_ROWS) / max(1, TOTAL))

print(f"Total rows available: {TOTAL}; sampling fraction used: {fraction:.6f}")

# 6) Sample and convert to pandas (this is the step that brings data to driver)
sampled_spark = spark_cluster_df.sample(fraction=fraction, seed=42)

# If sampling returns more than MAX_ROWS (rare), limit it:
sampled_spark = sampled_spark.limit(MAX_ROWS)

sample_pd = sampled_spark.toPandas()
print(f"Pandas sample shape: {sample_pd.shape}")

# 7) Safety check
if sample_pd.shape[0] == 0:
    raise ValueError("Sampled dataframe is empty. Reduce MAX_ROWS or check data availability.")

# 8) Local preprocessing with scikit-learn
X = sample_pd[cluster_features].values.astype(float)

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# 9) MiniBatchKMeans (lightweight, incremental)
n_clusters = 4            # change if you want more/less clusters
batch_size = 100          # mini-batch size
mbkm = MiniBatchKMeans(n_clusters=n_clusters, batch_size=batch_size, random_state=42)
mbkm.fit(X_scaled)
labels = mbkm.predict(X_scaled)

# 10) PCA for 2D visualization
pca = PCA(n_components=2, random_state=42)
X_pca = pca.fit_transform(X_scaled)

# 11) Compose results into a Pandas DF for plotting
viz_df = pd.DataFrame({
    'pc1': X_pca[:, 0],
    'pc2': X_pca[:, 1],
    'cluster': labels
})

# 12) Plot (seaborn)
plt.figure(figsize=(9,6))
sns.scatterplot(data=viz_df, x='pc1', y='pc2', hue='cluster', palette='tab10', s=50, alpha=0.8)
plt.title("Song Clusters — Musical & Emotional Similarity (sampled)")
plt.xlabel("PCA component 1")
plt.ylabel("PCA component 2")
plt.legend(title='Cluster', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()

display(viz_df)


# 13) Optional: attach cluster labels back to a small sample of songs (if you want song-level insight)
# If you have song identifiers in sentiment_df, sample them the same way and join labels locally:
# Example:
# sample_meta = sentiment_df.select('song','artist', *cluster_features).na.drop().sample(fraction=fraction, seed=42).limit(MAX_ROWS).toPandas()
# sample_meta['cluster'] = labels
# display(sample_meta.head(10))
```
